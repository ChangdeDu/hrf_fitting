## Step 1 alt: Package synth voxel data

## I need to sample the synth-voxels and place the data into a 2d matrix format with conditions (concatenated trn and val) vs. synth-voxels.

```python
layerNames = ['r1', 'r2', 'r3']

idxs = {}
synth_val_data = {}
synth_trn_data = {}
pruneNodes = 0.90;
totalNodes = 0;

for l in layerNames:
    valFrameCount = deepnet_val_feature_dict[l].shape[0]
    trnFrameCount = deepnet_trn_feature_dict[l].shape[0]
    nodeCount  = 1
    for i in range(1,len(deepnet_trn_feature_dict[l].shape)):
        nodeCount *= deepnet_trn_feature_dict[l].shape[i] 
       
    idxs[l] = np.arange(nodeCount)
    np.random.shuffle(idxs[l])

    keepNodes = int((1. - pruneNodes) * float(nodeCount))
    totalNodes += keepNodes

    idxs[l] = idxs[l][0:keepNodes]
    #print len(idxs[l])
    
    val_data = np.reshape(deepnet_val_feature_dict[l], (valFrameCount, nodeCount))
    synth_val_data[l] = val_data[:, idxs[l]]
    print "%s --> %s" % (val_data.shape, synth_val_data[l].shape)

    trn_data = np.reshape(deepnet_trn_feature_dict[l], (trnFrameCount, nodeCount))
    synth_trn_data[l] = trn_data[:, idxs[l]]
    print "%s --> %s" % (trn_data.shape, synth_trn_data[l].shape)
    #synth_trn_data = np.reshape(deepnet_trn_feature_dict[layerName], (trnFrameCount, nodeCount))
    #print synth_trn_data.shape

del val_data
del trn_data
print "-------------------------------"
print "total node count = %d" % totalNodes
```

```python
#concatenate selected voxels
val_data = np.ndarray((valFrameCount, 0))
trn_data = np.ndarray((trnFrameCount, 0))
start_idx = {};
stop_idx = {};

for l in layerNames:
    start_idx[l] = val_data.shape[1]
    val_data = np.concatenate((val_data, synth_val_data[l]), axis=1).astype('float32')    
    trn_data = np.concatenate((trn_data, synth_trn_data[l]), axis=1).astype('float32')    
    stop_idx[l] = val_data.shape[1]
    
print val_data.shape
print trn_data.shape
```

## Step 2: Load the stimuli data

```python
##allocate memory for stim
training_stim = np.zeros((Ttrn,S,S),dtype='float32')

##load training stim
cnt = 0
for sl in sorted(train_stim_files):
    this_h5 = h5py.File(sl,'r')
    this_train_stim = this_h5['stimTrn']
    this_num_stim = this_train_stim.shape[-1]
    training_stim[cnt:cnt+this_num_stim,:,:] = np.transpose(this_train_stim[:], [2,1,0])
    cnt += this_num_stim
    this_h5.close()
    
##load validation stim
val_h5 = h5py.File(val_stim_file,'r')
validation_stim = np.transpose(val_h5['stimVal'][:],[2,1,0])
val_h5.close()
```

### Construct the model space (not sure what you mean by that? Seems to be the constructing the model responses)

```python
log_act_func = lambda x: np.log(1+np.sqrt(x))
```

```python
##construct the model space
init_feat_dict = gfm.create_feature_maps(training_stim[0,np.newaxis,np.newaxis,:,:])
ms = model_space(init_feat_dict, rf, activation_function = log_act_func)
```

```python
##loop over training stimuli because feature maps for all training stim. > 48Gb
trn_mst = np.zeros((ms.receptive_fields.G, Ttrn, ms.D)).astype('float32')

num_chunks = 2
stim_dx = np.linspace(0,T-1,num=num_chunks+1, endpoint=True,dtype='int')

cnt = 0
for t in stim_dx[1:]:
    this_training_feature_dict = gfm.create_feature_maps(training_stim[cnt:cnt+t,np.newaxis,:,:])
    trn_mst[:,cnt:cnt+t,:] = ms.construct_model_space_tensor(this_training_feature_dict,normalize=False)
    cnt += t

##clear up memory
this_training_feature_dict = []

##normalize and save normalization constants
trn_mst = ms.normalize_model_space_tensor(trn_mst, save=True)
```

```python
##should work in one shot for because not too big
val_feature_dict = gfm.create_feature_maps(validation_stim[:,np.newaxis,:,:])

val_mst = ms.construct_model_space_tensor(val_feature_dict,normalize=True)

##Shouldn't we normalize and save normalization constants???
```

## Step 5: run that shit.

#### initialize the feature weights

```python
initial_feature_weights = 'zeros'
```

#### train the model!

```python
# trn_mst is the encoding model, i.e. the gabor wavelet activity.
# trn_voxel_data needs to be the layer output of the gsynet.

fvl,ffw,frf,beh = train_fwrf_model(trn_mst,
                 trn_voxel_data,
                 initial_feature_weights = initial_feature_weights,
                 voxel_binsize = nvox,
                 rf_grid_binsize=10,
                 learning_rate=10**(-7.0),
                 max_iters = 300,
                 early_stop_fraction=0.2,
                 report_every = 100)
```

#### prediction accuracy for all voxels

```python


##grab validation data
val_voxel_data = voxel_data[valIdx,0:nvox]
```

```python
##generate predictions
# pred = prediction_menu(val_mst, ffw[np.newaxis,:,:], rf_indices = frf) ##<<too big, choked. 


##generate predictions one voxel at a time
pred = np.zeros((Tval,nvox))
for v in range(nvox):  ##FIXED ! ?<<some kind of bug in training function, last voxel getting skipped...th
    pred[:,v] = np.squeeze(bigmult(val_mst[np.newaxis,frf[v],:,:],
                                   ffw[np.newaxis,:,v, np.newaxis]))
```

```python
##get correlation = prediction accuracy
val_cc = np.zeros(nvox) 
for v in range(nvox): 
    cc = pearsonr(val_voxel_data[:,v],pred[:,v])
    val_cc[v]=cc[0]
val_cc = np.nan_to_num(val_cc)
```

#### save model and val_cc

```python
saving_file = image_path + 'model_space_'+version_number+'.p'

ms.optimal_feature_weights = ffw
ms.optimal_rf_model = frf
ms.val_cc = val_cc
```

#### area-wise rf size vs. eccentricity

```python
## construct data table of rfs--indices will now be voxels
frf_sizes = ms.receptive_fields.rf_table.loc[frf,'deg_per_radius'].values
frf_eccentricities = ms.receptive_fields.rf_table.loc[frf, ['x_deg','y_deg']].apply(lambda row: np.sqrt(row['x_deg']**2+row['y_deg']**2),axis=1).values
named_voxels = [roi_names[int(v)] for v in roi_indicator]
```

```python
model_table = pd.DataFrame(data=list(frf_sizes), columns=['roi_size'])
model_table['eccentricity'] = frf_eccentricities
model_table['area'] = named_voxels
model_table['valcc'] = val_cc

model_table
```

```python
val_thresh = 0.25
size_means = {}
size_stds = {}
area_eccs = {}
for roi in roi_names:
    area_stats = model_table.loc[(model_table['area']==roi) & (model_table['valcc'] > val_thresh),['roi_size', 'eccentricity']]
    ecc_grp = area_stats.groupby(by='eccentricity')
    size_means[roi] = []
    size_stds[roi] = []
    area_eccs[roi] = []
    for name,grp in ecc_grp:
        area_eccs[roi].append(name)
        size_means[roi].append(grp.roi_size.mean())
        size_stds[roi].append(grp.roi_size.std())
    
```

```python
roi_colors = ['black', 'r', 'salmon', 'coral', 'darkslategray', 'teal','orangered', 'indigo' ]
ecc = np.sort(np.unique(model_table['eccentricity']))
line_list = []
fig = plt.figure(figsize=(10,10))
for r,roi in enumerate(roi_names):
    plt.plot(area_eccs[roi],size_means[roi], 'o', color=roi_colors[r])
    p = np.polyfit(area_eccs[roi],size_means[roi],1)
    l, = plt.plot(area_eccs[roi], np.polyval(p,area_eccs[roi]), color=roi_colors[r], label=roi, linewidth=4)
    line_list.append(l)
    
plt.legend(loc='upper left', ncol = 4)    
# plt.axis('equal')
plt.xlim([0,14])
plt.ylim([0,8])
```

#### layer x area weight distribution

```python
layer_names = ['p1', 'r1', 'p2', 'r2', 'r3']
weights = np.zeros(( len(ms.feature_depth.keys()), len(roi_names),))

for ii,roi in enumerate(roi_names):
    voxels_in_roi = roi_indicator == ii
    layer_cnt = 0
    for layer in layer_names:
        idx = ms.feature_indices[layer]
        weights[layer_cnt,ii] = np.mean(ms.optimal_feature_weights[idx,:][:,voxels_in_roi])
        layer_cnt += 1

plt.pcolor(weights, cmap='hot')
plt.ylabel(layer_names)
plt.xlabel(roi_names)
        
```
